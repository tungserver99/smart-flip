```python
from pathlib import Path
from src.smart_flip.evaluation.lm_eval import LMEvalHarnessRunner

root = Path("/content/drive/Othercomputers/My Computer/Google Drive Offline/PHD/Quantization/smart-flip")

flip_candidates = sorted((root / "results/models/awq_flip").glob("awq_flip_run*"))
raw_candidates = sorted((root / "results/models/awq_raw").glob("awq_raw_run*"))

print("flip candidates:")
for p in flip_candidates:
    print(" -", p)

print("raw candidates:")
for p in raw_candidates:
    print(" -", p)

flip_path = flip_candidates[-1]
raw_path = raw_candidates[-1]

print("\nchosen flip:", flip_path, "exists:", flip_path.exists())
print("chosen raw :", raw_path, "exists:", raw_path.exists())

runner = LMEvalHarnessRunner(
    tasks=["lambada_openai"],
    output_dir=str(root / "results/eval/tmp"),
    run_name="tmp_lambada_saved_models",
)

results = runner.run({
    "awq_flip": str(flip_path.resolve()),
    "awq_raw": str(raw_path.resolve()),
})

results
```
